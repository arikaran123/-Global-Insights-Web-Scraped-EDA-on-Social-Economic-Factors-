```python
data_info2 = []

def GetData(tag):
    url2 = f'https://data.worldbank.org/country/{tag}?view=chart'
    response2 = requests.get(url2)
    soup2 = BeautifulSoup(response2.text, 'html.parser')
    yes = soup2.find_all('div', {'class': 'indicator-item__data-info'})
    country_data = []  
    counter = 0
    for row in yes:
        no = row.find_all(attrs={'span': False})
        row_data = []  
        for step in no:
            text1 = step.get_text()
            row_data.append(text1)
            counter += 1
            if counter >= 13:  
                break
            country_data.append(row_data)  
        if counter >= 13: 
            break
    print(f'Data for {tag}: {country_data}') 
    name = soup2.find('div', {'class': 'cardheader'})
    country = name.find(attrs= {'div': False})
    place = country.get_text()
    country_data.append(place)
    return country_data
```

```python
countries = ['China', 'Canada', 'India', 'Mexico', 'United States', 'Finland', 'Argentina', 'Netherlands', 'Pakistan']

# Initialize an empty list to store data for all countries
all_data = []

# Loop through each country and fetch data
for country in countries:
    country_data = GetData(country)
    all_data.append(country_data)

# Create a DataFrame using the collected data and new_headers
new_row_df = pd.DataFrame(all_data, columns=new_headers)

# Print the new DataFrame
new_row_df
```

```python
numeric_columns = ['Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population)',
                   'Life expectancy at birth, total (years)',
                   'Population, total',
                   'Population growth (annual %)',
                   'Net migration',
                   'Human Capital Index (HCI) (scale 0-1)',
                   'GDP (current US$)',	
                   'GDP per capita (current US$)',
                   'GDP growth (annual %)',
                   'Unemployment, total (% of total labor force) (modeled ILO estimate)',
                   'Inflation, consumer prices (annual %)',
                   'Personal remittances, received (% of GDP)']

# Remove square brackets from numeric columns only
new_row_df[numeric_columns] = new_row_df[numeric_columns].applymap(lambda x: x[0])

# Print the updated DataFrame without affecting 'Country' column
new_row_df
```
